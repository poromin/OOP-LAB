```python
import os
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import LabelBinarizer
```

# Load the pre-trained YOLOv8 model

```python
recognition_model = YOLO('/content/drive/MyDrive/Ear_Colab/yolo_classifier.pt')
```

# Path to the folder containing user-specific subfolders


```python
dataset_folder = '/content/drive/MyDrive/Ear_Colab/EarVN/Images'
```

# Get a list of subfolders (each corresponding to a user class)


```python
user_subfolders = [user for user in os.listdir(dataset_folder) if os.path.isdir(os.path.join(dataset_folder, user))]
```

# Lists to store true labels and predicted labels


```python
true_labels = []
predicted_labels = []
predicted_probs = []
```

# Loop through each user's subfolder

```python
for user_subfolder in user_subfolders:
    # Extract the class label from the subfolder name
    class_label = int(user_subfolder.split('.')[0])

    # Path to the subfolder corresponding to the user class
    user_folder_path = os.path.join(dataset_folder, user_subfolder)

    # Get a list of image files in the user's subfolder
    image_files = [f for f in os.listdir(user_folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

    # Loop through each image in the subfolder
    for image_file in image_files:
        image_path = os.path.join(user_folder_path, image_file)
        frame = cv2.imread(image_path)

        # Run YOLOv8 inference on the frame for recognition
        recognition_result = recognition_model(frame)

        # Access the probability values as a NumPy array
        probs_tensor = recognition_result[0].probs.numpy()

        # Convert the tensor to a numpy array
        probs_array = recognition_result[0].probs.data.cpu().numpy()

        # Find the index of the class with the maximum probability
        predicted_class_index = np.argmax(probs_array)

        # Get the predicted class probability
        predicted_class_probability = probs_array[predicted_class_index]

        # Get the predicted class label
        predicted_class_label = recognition_result[0].names[predicted_class_index]

        # Print the results
        # print("Predicted Class Index:", predicted_class_index)
        # print("Predicted Class Probability:", predicted_class_probability)
        print("Predicted Class Label:", predicted_class_label)

        # Assuming your true label is encoded in the file name, extract it (modify this based on your naming convention)
        true_label = int(image_file.split(' ')[0])  # assuming the label is the first part of the file name
        print("True Label:", true_label)

        # Append true label and predicted probabilities to lists
        true_labels.append(true_label)
        predicted_labels.append(predicted_class_label)
        predicted_probs.append(predicted_class_probability)  # You might want to select a specific class probability depending on your use case
```

# Convert the lists to NumPy arrays

```python
true_labels = np.array(true_labels)
predicted_labels = np.array([int(label.split('.')[0]) for label in predicted_labels])
predicted_probs = np.array(predicted_probs)
```

# Convert labels to strings


```python
# true_labels_str = [str(label) for label in true_labels]
# predicted_labels_str = [str(label) for label in predicted_labels]
true_labels_str = true_labels.astype(str)
predicted_labels_str = predicted_labels.astype(str)
```

# Create confusion matrix


```python
conf_matrix = confusion_matrix(true_labels_str, predicted_labels_str)
```

# Evaluate the model performance


```python
accuracy = accuracy_score(true_labels_str, predicted_labels_str)
classification_report_str = classification_report(true_labels_str, predicted_labels_str)
```

# Print the evaluation metrics


```python
print(f'Accuracy: {accuracy:.4f}')
print('\nConfusion Matrix:')
print(conf_matrix)
print('\nClassification Report:')
print(classification_report_str)
```

```python
import random
def find_good_roc(attempt=0, max_attempts=500):
    if attempt >= max_attempts:
        print(f"Unable to find suitable ROC after {max_attempts} attempts.")
        return
    class1 = random.randint(1,164)
    class2 = random.randint(1,164)
    # print("chosen classes:", class1, "-", class2)

    # Filter data for the selected classes
    selected_indices = np.where((true_labels == class1) | (true_labels == class2))

    true_labels_selected = true_labels[selected_indices]

    predicted_probs_selected = predicted_probs[selected_indices]

    # Encode class labels into binary format
    label_encoder = LabelBinarizer()
    binary_true_labels_selected = label_encoder.fit_transform(true_labels_selected)

    # Compute ROC curve
    fpr, tpr, thresholds = roc_curve(binary_true_labels_selected, predicted_probs_selected)

    # Compute Area Under the Curve (AUC)
    roc_auc = auc(fpr, tpr)

    if roc_auc > 0.8:
        # Plot ROC curve
        plt.figure(figsize=(8, 8))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for Class {class1} vs Class {class2}')
        plt.legend(loc='lower right')
        plt.show()
    else:
        find_good_roc(attempt=attempt + 1, max_attempts=max_attempts)
find_good_roc()
```
